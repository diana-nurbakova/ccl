## Conclusions

The four experiments validate four independent CCL mechanisms:

1. **Differential scaffolding by claim type** — not all errors are equally visible (Exp A)
2. **Structural enforcement of evaluation** — not voluntary critical thinking (Exp B)
3. **Process-level reflection over time** — engagement decays without it (Exp C)
4. **Sustained scaffolding over long timescales** — decay continues over months (Exp D)

Together they support the CCL's claim that critical engagement with AI requires structured, sustained intervention — not just instructions to "think critically."
