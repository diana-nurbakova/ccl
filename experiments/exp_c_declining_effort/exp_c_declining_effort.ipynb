# Experiment C: Declining Effort Across Sessions

## CCL Prediction
Without process-level reflection (Stage 3), critical engagement declines across sessions.

## Data
**Bastani et al. (2025)**. 451 students with ≥3 of 4 sessions. Within-student trajectory analysis.

## 1. Load and Prepare Data

```python
from experiments.shared.data_acquisition import load_bastani_conversations
from experiments.exp_c_declining_effort.analysis import build_session_metrics, filter_repeat_students

conv_df = load_bastani_conversations()
session_df = build_session_metrics(conv_df)
filtered = filter_repeat_students(session_df, min_sessions=3)
print(f"{filtered['username'].nunique()} students with ≥3 sessions")
```

## 2. Session-Level Summary

```python
from experiments.exp_c_declining_effort.analysis import session_level_summary

summary = session_level_summary(filtered)
summary
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
for i, metric in enumerate(['passive_rate', 'active_rate', 'evaluative_rate']):
    for treatment in summary['treatment'].unique():
        t_data = summary[summary['treatment'] == treatment]
        axes[i].plot(t_data['session_id'], t_data[metric], marker='o', label=treatment)
    axes[i].set_title(metric.replace('_', ' ').title())
    axes[i].set_xlabel('Session')
    axes[i].legend()
fig.suptitle('Engagement Trajectories Across Sessions', fontweight='bold')
fig.tight_layout()
plt.show()
```

## 4. Interpretation

**What supports the CCL prediction:**
- Passive rates increase significantly in both conditions.
- Active engagement declines significantly for GPT Tutor students.
- Even structured scaffolding does not fully prevent disengagement over time.

**What complicates the prediction:**
- Evaluative rate increases for GPT Base — a learning curve effect.
- The 4-session window may be too short for dramatic decay.

**Implication**: The CCL's Stage 3 (retrospective/process reflection) is needed not because evaluation drops to zero, but because passivity increases even under scaffolding.
